```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
#Sending request to webpage

url = "https://www.aldi.us/"
```

```python
#send a rrequest to get a text fo this webpage
res = requests.get(url).text
res
```

```python
#Making sense of the html parts
soup = BeautifulSoup(res, 'html.parser')
```

```python
print(soup.prettify())
```

```python
#having a title is very important, metacontent, descritpion aswell.
#we will look for all the descriptons, and than save all the warnings, titles, alt and etc

bad = []
good = []
keywords = []
```

```python
#title

title = soup.find('title').text
if title:
    good.append(f"Title Exists: {title}")
else:
    bad.append("No Title!")

#Meta

meta_d = soup.find('meta', attrs={'name':'description'})['content']
if title:
    good.append(f"Meta Description Exists: {meta_d}")
else:
    bad.append("No Meta Description!")
#Headings
hs = ["h1", 'h2', 'h3']
h_tags = []
for h in soup.find_all(hs):
    good.append(f"{h.name} --->{h.text.strip()}")
    h_tags.append(h.name)
if 'h1' not in h_tags:
    bad.append("No H1 found!")

#Image ALt(Images do not have an alt attribute)
#below is showing that if there is no images than add to the 'bad' list
for i in soup.find_all('img', alt=''):
    bad.append(f"No Alt: {i}")
print(good)
print('------------')
print(bad)
```
